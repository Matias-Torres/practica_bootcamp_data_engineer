```sql
%sql

--Analisis de valores Nulos por cada columna
WITH total AS (

SELECT 
  COUNT(*) AS total_registros
FROM bootcamp_matias.raw.propiedades_bronze

),

nulos_por_columna AS (

SELECT 
  'precio' AS columna, 
  COUNT(*) AS nulos
FROM bootcamp_matias.raw.propiedades_bronze
WHERE 
  precio IS NULL

UNION ALL
SELECT 
  'ubicacion', 
  COUNT(*) 
FROM bootcamp_matias.raw.propiedades_bronze 
WHERE 
  ubicacion IS NULL

UNION ALL

SELECT
 'numero', 
 COUNT(*) 
FROM bootcamp_matias.raw.propiedades_bronze 
WHERE 
  numero IS NULL

UNION ALL

SELECT 
  'calle', 
  COUNT(*) 
FROM bootcamp_matias.raw.propiedades_bronze 
WHERE 
  calle IS NULL

UNION ALL

SELECT 
'expensas',
 COUNT(*) 
FROM bootcamp_matias.raw.propiedades_bronze 
WHERE 
  expensas IS NULL

UNION ALL

SELECT 
  'tipo_de_operacion', 
  COUNT(*) 
FROM bootcamp_matias.raw.propiedades_bronze 
WHERE 
  tipo_de_operacion IS NULL

UNION ALL

SELECT 
  'moneda', 
  COUNT(*) 
FROM bootcamp_matias.raw.propiedades_bronze 
WHERE 
  moneda IS NULL

UNION ALL

SELECT 
  'ambientes', 
  COUNT(*) 
FROM bootcamp_matias.raw.propiedades_bronze 
WHERE 
  ambientes IS NULL

UNION ALL

SELECT 
  'metros_cuadrados_totales', 
  COUNT(*) 
FROM bootcamp_matias.raw.propiedades_bronze 
WHERE 
  metros_cuadrados_totales IS NULL

UNION ALL

SELECT 
  'metros_cuadrados_cubiertos',
  COUNT(*) 
FROM bootcamp_matias.raw.propiedades_bronze 
WHERE 
  metros_cuadrados_cubiertos IS NULL

UNION ALL

SELECT 
  'orientacion_cardinal', 
  COUNT(*) 
FROM bootcamp_matias.raw.propiedades_bronze 
WHERE 
  orientacion_cardinal IS NULL

UNION ALL

SELECT 
  'orientacion_inmueble', 
  COUNT(*) 
FROM bootcamp_matias.raw.propiedades_bronze 
WHERE 
  orientacion_inmueble IS NULL

UNION ALL

SELECT 
  'piso', 
  COUNT(*) 
FROM bootcamp_matias.raw.propiedades_bronze 
WHERE 
  piso IS NULL

UNION ALL

SELECT 
  'cochera', 
  COUNT(*) 
FROM bootcamp_matias.raw.propiedades_bronze 
WHERE 
  cochera IS NULL

UNION ALL

SELECT 
  'antiguedad', 
  COUNT(*) 
FROM bootcamp_matias.raw.propiedades_bronze 
WHERE 
  antiguedad IS NULL

UNION ALL

SELECT 
  'estado', 
  OUNT(*) 
FROM bootcamp_matias.raw.propiedades_bronze 
WHERE 
  estado IS NULL

UNION ALL

SELECT 
  'tipo_vendedor', 
  COUNT(*) 
FROM bootcamp_matias.raw.propiedades_bronze 
WHERE
   tipo_vendedor IS NULL

)

SELECT
    columna,
    nulos,
    ROUND(nulos * 100.0 / total.total_registros, 2) AS pct_nulos
FROM nulos_por_columna
CROSS JOIN total
--WHERE ROUND(nulos * 100.0 / total.total_registros, 2) > 50
ORDER BY pct_nulos DESC;

/*
Se realizó un análisis de completitud de datos sobre el dataset para identificar columnas con alto porcentaje de valores faltantes y evaluar su impacto en el modelo analítico.

Columnas con alto porcentaje de valores nulos (>50%), indica baja disponibilidad de informacion en el dataset. Estas columnas no se consideran criticas para el modelo analítico, por lo que se decidio posibles imputaciones o excluirla del analisis

Columnas con nulos moderados (5 -20%), posible regla de imputacion, filtrado de registros o transforamciones derivadas

Columna principales para el analisis:
precio
tipo_de_operacion
ubicacion
moneda
*/
```
