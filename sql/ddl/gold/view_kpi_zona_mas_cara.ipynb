```sql
%sql
DROP VIEW IF EXISTS bootcamp_matias.gold.kpi_zona_caras_alquiler;

CREATE VIEW bootcamp_matias.gold.kpi_zona_caras_alquiler
COMMENT 'KPI Top 10 de Zona mas caras'
AS
SELECT
DISTINCT
  B.zona,
  A.ambientes,
  ROUND(AVG(A.precio),2) AS precio_promedio,
  ROUND(AVG(A.precio_por_m2),2) AS precio_por_m2_promedio,
  CURRENT_TIMESTAMP() AS _refresh_timestamp
FROM bootcamp_matias.gold.fact_propiedades A
LEFT JOIN bootcamp_matias.gold.dim_zona B
  ON A.zona_id = B.zona_id
LEFT JOIN bootcamp_matias.gold.dim_tipo_operacion C
  ON A.tipo_operacion_id = C.tipo_operacion_id
WHERE 
  C.tipo_operacion = 'alquiler'
  AND A.precio > 0
  AND A.precio_por_m2 > 0
  AND A.ambientes > 0 
GROUP BY 
  B.zona, A.ambientes
ORDER BY 
  precio_promedio DESC
LIMIT 10
```

```sql
%sql
SELECT * FROM bootcamp_matias.gold.kpi_zona_caras_alquiler
```
