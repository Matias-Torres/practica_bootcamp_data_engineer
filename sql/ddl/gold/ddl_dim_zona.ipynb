```sql
%sql
DROP TABLE IF EXISTS bootcamp_matias.gold.dim_zona;

CREATE TABLE bootcamp_matias.gold.dim_zona(
  zona_id BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1 ),
  zona STRING NOT NULL,
  ciudad STRING NOT NULL,
  provincia STRING DEFAULT 'Buenos Aires',
  fecha_inicio DATE NOT NULL,
  fecha_final DATE NOT NULL,
  es_actual BOOLEAN NOT NULL,
  _created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP()
)
USING DELTA
TBLPROPERTIES (
  'delta.feature.allowColumnDefaults' = 'supported'
)
COMMENT 'Dimension Zona - Star Schema - SCD Type 2'
```
