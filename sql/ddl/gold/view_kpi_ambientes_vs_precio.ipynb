```sql
%sql
DROP VIEW IF EXISTS bootcamp_matias.gold.kpi_ambientes_vs_precio;

CREATE VIEW bootcamp_matias.gold.kpi_ambientes_vs_precio
AS
SELECT
    A.ambientes,

    COUNT(*) AS cantidad,

    ROUND(AVG(A.precio),2) AS precio_promedio,
    ROUND(AVG(A.precio_por_m2),2) AS precio_m2_promedio

FROM bootcamp_matias.gold.fact_propiedades A
LEFT JOIN bootcamp_matias.gold.dim_tipo_operacion B 
  ON A.tipo_operacion_id = B.tipo_operacion_id
WHERE 
  B.tipo_operacion = 'alquiler'
  AND A.moneda = 'ARS'
  AND A.precio > 0
  AND A.precio_por_m2 > 0
  AND A.ambientes > 0
GROUP BY 
  A.ambientes
ORDER BY 
  A.ambientes ASC 
```
