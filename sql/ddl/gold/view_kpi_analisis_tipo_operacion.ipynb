```sql

DROP VIEW IF EXISTS bootcamp_matias.gold.kpi_analisis_tipo_operacion;

CREATE VIEW bootcamp_matias.gold.kpi_analisis_tipo_operacion
AS
SELECT

  B.tipo_operacion,

  COUNT(*) AS cantidad_publicaciones,

  ROUND(AVG(A.precio),2) AS precio_promedio,
  ROUND(AVg(A.precio_por_m2),2) AS precio_m2_promedio

FROM bootcamp_matias.gold.fact_propiedades A 
LEFT JOIN bootcamp_matias.gold.dim_tipo_operacion B
  ON A.tipo_operacion_id = B.tipo_operacion_id
WHERE 
  A.moneda = 'ARS'
  AND A.precio > 0 
  AND A.precio_por_m2 > 0
  AND B.tipo_operacion != 'venta_alquiler'
GROUP BY 
  B.tipo_operacion
ORDER BY 
  cantidad_publicaciones DESC
```
