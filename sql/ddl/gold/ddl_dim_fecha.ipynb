```sql
%sql
CREATE OR REPLACE TABLE bootcamp_matias.gold.dim_fecha
COMMENT 'Tabla de dimension de fechas - Star Schema - Sin SCD'

WITH fechas AS (
  SELECT SEQUENCE(
    DATE_ADD(current_date(), -5 * 365),
    DATE_ADD(current_date(), 5 * 365),
    INTERVAL 1 DAY
  ) AS fechas_array
)
SELECT 
  CAST(DATE_FORMAT(fecha, 'yyyMMdd') AS BIGINT) AS id_fecha,
  explode(fechas_array) AS fecha,
  YEAR(fecha) AS anio,
  MONTH(fecha) AS mes,
  DATE_FORMAT(fecha, 'MMMM') AS nombre_mes,
  QUARTER(fecha) AS trimestre,
    CASE DAYOFWEEK(fecha)
      WHEN 1 THEN 'Domingo'
      WHEN 2 THEN 'Lunes'
      WHEN 3 THEN 'Martes'
      WHEN 4 THEN 'Miercoles'
      WHEN 5 THEN 'Jueves'
      WHEN 6 THEN 'Viernes'
      WHEN 7 THEN 'Sabado'
    END AS dia_semana,
  DAYOFWEEK(fecha) IN (1,7)  as es_fin_de_semana,
  CURRENT_TIMESTAMP() AS _created_at
FROM fechas
```
