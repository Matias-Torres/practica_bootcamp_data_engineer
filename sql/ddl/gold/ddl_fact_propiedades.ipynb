```sql
%sql
DROP TABLE IF EXISTS bootcamp_matias.gold.fact_propiedades;

CREATE TABLE bootcamp_matias.gold.fact_propiedades(
propiedades_id BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
zona_id BIGINT,
tipo_operacion_id BIGINT,
estado_id BIGINT,
precio DECIMAL(10,2),
moneda STRING,
expensas DECIMAL (10,2),
ambientes INT,
metros_cuadrados_totales DECIMAL(10,2),
metros_cuadrados_cubiertos DECIMAL(10,2) ,
antiguedad INT,
precio_por_m2 DECIMAL(10,2),
url STRING,
_refresh_timestamp TIMESTAMP
)
USING DELTA
TBLPROPERTIES 
    (
    'delta.feature.allowColumnDefaults' = 'supported',
    'delta.autoOptimize.optimizeWrite' = 'true',
    'delta.autoOptimize.autoCompact' = 'true'
    )
COMMENT 'Tabla de hechos de propiedades - Star Schema';
```
