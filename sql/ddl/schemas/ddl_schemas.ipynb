```sql
%sql

-- PASO 1: Crear catálogo
CREATE CATALOG IF NOT EXISTS bootcamp_matias;

USE CATALOG bootcamp_matias;

SHOW CATALOGS;
```

```sql
-- PASO 2: Crear Esquemas RAW-SILVER-GOLD
CREATE SCHEMA IF NOT EXISTS bootcamp_matias.raw
COMMENT 'Esquema para datos crudos sin procesar (Bronze layer)';

CREATE SCHEMA IF NOT EXISTS bootcamp_matias.silver
COMMENT 'Esquema para datos limpios (Silver Layer)';

CREATE SCHEMA IF NOT EXISTS bootcamp_matias.gold
COMMENT 'Capa Gold: Datos agregados y métricas para negocio';

SHOW SCHEMAS;
```

```sql
-- PASO 3: Crear el Volume para archivos
CREATE VOLUME IF NOT EXISTS bootcamp_matias.raw.archivos
COMMENT 'Volume para almacenar archivos CSV crudos';

SHOW VOLUMES IN bootcamp_matias.raw;
```

```sql
-- PASO 4: Verificar que el archivo se subió correctamente
LIST '/Volumes/bootcamp_matias/raw/archivos';
```
