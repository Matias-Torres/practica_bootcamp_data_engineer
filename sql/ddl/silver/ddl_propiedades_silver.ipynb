```sql
DROP TABLE IF EXISTS bootcamp_matias.silver.propiedades;

CREATE TABLE bootcamp_matias.silver.propiedades(
    -- Identificador (auto-incremental)
    propiedad_id BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1) COMMENT 'ID único de la propiedad (auto-generado)',
    
    -- Ubicación
    zona STRING COMMENT 'Zona/region de la propiedad',
    ciudad STRING COMMENT 'Ciudad de la propiedad',

    -- Características principales
    tipo_operacion STRING COMMENT 'alquiler, venta, venta_alquiler',
    precio DECIMAL(10,2) COMMENT 'Precio de la propiedad',
    moneda STRING COMMENT 'USD o ARS',
    expensas DECIMAL(10,2) COMMENT 'Expensas mensuales',
    
    -- Dimensiones
    ambientes INT COMMENT 'Cantidad de ambientes 1-6 (NULL, no disponible)',
    metros_cuadrados_totales DECIMAL(10,2) COMMENT 'Superficie total',
    metros_cuadrados_cubiertos DECIMAL(10,2) COMMENT 'Superficie cubierta',
    
    -- Características adicionales
    antiguedad INT COMMENT 'Años de antigüedad (999, no disponible)',
    estado STRING COMMENT 'Estado de la propiedad',
    url STRING COMMENT 'URL de la publicacion',
    
    -- Métricas calculadas
    precio_por_m2 DECIMAL(10,2) COMMENT 'Precio por metro cuadrado',
    
    -- Metadata de procesamiento
    _source_table STRING DEFAULT 'bootcamp_matias.raw.propiedades_bronze' COMMENT 'Tabla origen',
    _processing_timestamp TIMESTAMP DEFAULT CURRENT_TIMESTAMP() COMMENT 'Timestamp de procesamiento'
)
USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported')
COMMENT 'Propiedades inmobiliarias - Capa Silver (datos limpios y validados)';
```
