```sql
%sql
INSERT INTO bootcamp_matias.gold.dim_tipo_operacion (tipo_operacion, _created_at)
SELECT
 DISTINCT
  A.tipo_operacion,
  CURRENT_TIMESTAMP()
FROM bootcamp_matias.silver.propiedades A

WHERE NOT EXISTS(
  SELECT 1
  FROM bootcamp_matias.gold.dim_tipo_operacion B 
  WHERE 
    A.tipo_operacion = B.tipo_operacion
);
```
