```sql
WITH diagnostico_calidad AS (
    SELECT
        COUNT(*) AS total_registros,
        
        -- Precios problemáticos
        COUNT(CASE WHEN precio IS NULL THEN 1 END) as precio_nulo,
        COUNT(CASE WHEN precio <= 0 THEN 1 END) AS precio_invalido,
        
        -- Metros cuadrados
        COUNT(CASE WHEN metros_cuadrados_totales IS NULL OR metros_cuadrados_totales <= 0 THEN 1 END) AS m2_invalido,
        
         -- Ambientes
        COUNT(CASE WHEN ambientes IS NULL OR ambientes <= 0 THEN 1 END) as ambientes_invalido,

        -- Estado
        COUNT(CASE WHEN estado IS NULL OR estado = '' THEN 1 END) as estado_invalido,

        -- Antigüedad con valor 999
        COUNT(CASE WHEN cast(antiguedad AS FLOAT) = 999 OR antiguedad = '999' THEN 1 END) as antiguedad_placeholder,
        
        -- Campos obligatorios vacíos
        COUNT(CASE WHEN zona IS NULL OR zona = '' THEN 1 END) as zona_vacia,
        COUNT(CASE WHEN tipo_operacion IS NULL OR tipo_operacion = '' THEN 1 END) AS tipo_operacion_vacio
        
    FROM bootcamp_matias.silver.propiedades
)
SELECT 
    total_registros,
    precio_nulo,
    precio_invalido,
    m2_invalido,
    ambientes_invalido,
    estado_invalido,
    antiguedad_placeholder,
    zona_vacia,
    tipo_operacion_vacio,
    ROUND((precio_nulo + precio_invalido) * 100.0 / total_registros, 2) as pct_precio_problematico
FROM diagnostico_calidad;
```
